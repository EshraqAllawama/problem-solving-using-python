```python
def Math(operator,N1,N2):
  match operator:
    case "+": return N1+N2
    case "-": return N1-N2
    case "*": return N1*N2
    case "//": return N1//N2




```

```python
def to_Alternating_Case(st):
  for i in st:
    if ord(i) > 64 and ord(i) < 91:
      st =  st.replace(i,chr(ord(i) + 32))
    elif ord(i) > 96 and ord(i) < 123:
      st =  st.replace(i,chr(ord(i) - 32))
  return st
```

```python
print(to_Alternating_Case("HeLlo"))
print(to_Alternating_Case("hElLO"))
print(to_Alternating_Case("HELLO"))
print(to_Alternating_Case("hello"))
print(to_Alternating_Case("1234"))
print(to_Alternating_Case("HeLl234o3243"))
print(to_Alternating_Case("Hi,My Name is Eshraq and Ma age is 20 !!."))
```

```python
def Amer_to_Brit(floor):
  if floor>0:
    if floor>12:
      return floor-2
    else:
      return floor-1
  else:
    return floor
```

```python
print(Amer_to_Brit(1))
print(Amer_to_Brit(0))
print(Amer_to_Brit(5))
print(Amer_to_Brit(15))
```
